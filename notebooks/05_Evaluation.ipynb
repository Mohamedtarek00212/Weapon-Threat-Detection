# Evaluation

Evaluate a trained checkpoint against the prepared test partition.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))
from ultralytics import YOLO
from weapon_threat_detection.artifacts import utc_timestamp, write_json
from weapon_threat_detection.device import select_device

device = select_device()
data_yaml = ROOT / 'processed' / 'yolo_v1' / 'data.yaml'
weights_path = ROOT / 'runs' / 'REPLACE_WITH_RUN' / 'weights' / 'best.pt'
if not weights_path.exists(): raise FileNotFoundError(f'Update weights_path before evaluation: {weights_path}')
metrics = YOLO(weights_path).val(data=str(data_yaml), split='test', device=device.name)
report = write_json(ROOT / 'reports' / f'evaluation_{utc_timestamp()}.json', {'weights': str(weights_path), 'device': device.__dict__, 'metrics': str(metrics)})
print(report)
```
